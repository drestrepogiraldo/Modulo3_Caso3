### Gráfico de Barras

Los gráficos de barras son posiblemente la visualización de datos más simple. Asignan categorías a números con el fin de mostrar comparaciones numéricas entre categorías
Se observa por medio de columnas verticales cuya cantidad no debe ser muy grande. Representan una estimación de la tendencia central de una variable numérica con la altura de cada rectángulo y proporciona alguna indicación de la incertidumbre en torno a esa estimación utilizando barras de error.

### Gráficos de líneas

Muestran cambios en una variable respuesta sobre un intervalo continuo o un lapso de tiempo. Es posible comparar entre diferentes subgrupos de datos, pero se recomienda no tener más de 5 líneas. También es muy útil en el caso de tener muchas categorías. 
Los gráficos de líneas tienen una debilidad importante: a diferencia de los gráficos de barras, no son apropiados para datos categóricos nominales. Mientras que los gráficos de barras distinguen entre cada "tipo" de punto, los gráficos de líneas los mezclan. Entonces, un gráfico de líneas impone un orden a los valores en el eje horizontal, y el orden no tendrá sentido con algunos datos.

### Diagramas de Caja

Un diagrama de caja (o diagrama de caja y bigotes) muestra la distribución de datos cuantitativos de una manera que facilita las comparaciones entre variables o entre niveles de una variable categórica. El cuadro muestra los cuartiles del conjunto de datos, mientras que los bigotes se extienden para mostrar el resto de la distribución, excepto los puntos que se determinan como "valores atípicos" mediante un método que es una función del rango entre cuartiles.

### Diagramas de Violín

Un diagrama de violín juega un papel similar al de un diagrama de caja y bigotes. Muestra la distribución de datos cuantitativos a través de varios niveles de una (o más) variables categóricas de modo que esas distribuciones puedan compararse. A diferencia de un diagrama de caja, en el que todos los componentes del diagrama corresponden a puntos de datos reales, el diagrama de violín presenta una estimación de la densidad del núcleo de la distribución subyacente.

#### Ejemplo 7

Realizar un gráfico de violín para la cantidad de compras Web (`NumWebPurchases`) para la columna `Education`. 

```
sns.violinplot(x = 'Education', y = 'NumWebPurchases' ,data = df_customers) 
plt.ylim(-3,15)
```

```
sns.violinplot(x = 'Education', y = 'NumWebPurchases' ,hue="Kidhome",data = df_customers) 
plt.ylim(-3,15)
```
